```python
import math
import random
from math import sqrt
from matplotlib import pyplot as plt

count = 0
def getx(x, y, theta):
    return math.cos(theta)* x - math.sin(theta)*y
def gety(x, y, theta):
    return math.sin(theta)* x + math.cos(theta)*y
def get_colors():
    global count
    count += 1
    a = [round(float(random.random()), 6) for _ in range(3)]
    return '{}f, {}f, {}f,'.format(a[0], a[1], a[2])

def get_XY(a=0.5, n=10, hexa=False):
    X = []
    Y = []
    for i in range(n+1):
        x = round(float(a * math.cos(2 * math.pi * i /n)),6)
        y = round(float(a * math.sin(2 * math.pi * i /n)),6)
        X.append(x)
        Y.append(y)
        
    if hexa:
        for i in range(n):
            j = i + 1
            if i == n-1:
                j = 0
            z = round(sqrt(3)*a/2, 6)
            print('{}f, {}f, {}f, {}'.format(X[i], Y[i], 0.0, get_colors()))
            print('{}f, {}f, {}f, {}'.format(X[j], Y[j], 0.0, get_colors()))
            print('0.0f, 0.0f, {}, {}'.format(z, get_colors()))
            
            print('{}f, {}f, {}f, {}'.format(X[i], Y[i], 0.0, get_colors()))
            print('{}f, {}f, {}f, {}'.format(X[j], Y[j], 0.0, get_colors()))
            print('0.0f, 0.0f, {}, {}'.format(-z, get_colors()))

            
        return
    for i in range(n):
        j = i + 1
        if i == n-1:
            j = 0
        print('{}f, {}f, 0.0f, {}'.format(X[i], Y[i], get_colors()))
        print('{}f, {}f, 0.0f, {}'.format(X[j], Y[j], get_colors()))
        print('0.0f, 0.0f, 0.0f, {}'.format(get_colors()))
        
        print('{}f, {}f, {}f, {}'.format(X[i], Y[i], a, get_colors()))
        print('{}f, {}f, {}f, {}'.format(X[j], Y[j], a, get_colors()))
        print('0.0f, 0.0f, {}f, {}'.format(a, get_colors()))
        
        print('{}f, {}f, 0.0f, {}'.format(X[i], Y[i], get_colors()))
        print('{}f, {}f, 0.0f, {}'.format(X[j], Y[j], get_colors()))
        print('{}f, {}f, {}f, {}'.format(X[j], Y[j], a, get_colors()))
        
        print('{}f, {}f, {}f, {}'.format(X[i], Y[i], a, get_colors()))
        print('{}f, {}f, {}f, {}'.format(X[j], Y[j], a, get_colors()))
        print('{}f, {}f, 0.0f, {}'.format(X[i], Y[i], get_colors()))
#         print()
        
    plt.plot(X, Y)
    
    
```

```python
get_XY(n=10, hexa=False)
```
